```python
leaderlist = pd.read_csv('C:/Users/HP/Downloads/leaderlist_2_21.csv', encoding = "ISO-8859-1")
```

```python
#выбрали нужные колонки из датасета
leaderlist = leaderlist[['ccode', 'stateabb', 'leader', 'syear', 'l_year', 'militarycareer']]
```

```python
#склеили по коду страны, два датасета(режимы и лидеры)
general = regime_list.merge(leaderlist, how = 'outer' , left_on='cowcode', right_on='ccode')
```
